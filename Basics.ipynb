```python
import csv

f = open("guns.csv","r")
data = list(csv.reader(f))
data[:5]
```

```python
headers = data[0]
data = data[1:]
print(headers)
print(data[:5])
```

```python
years = [year[1] for year in data]
years
```

```python
year_counts = {}

for i in years:
    if i not in year_counts:
        year_counts[i] = 1
    else:
        year_counts[i] += 1
```

```python
year_counts
```

```python
import datetime
dates = [datetime.datetime(year=int(i[1]), month=int(i[2]), day=1) for i in data]
```

```python
dates
```

```python
date_counts = {}

for row in dates:
    if row not in date_counts:
        date_counts[row] = 1
    else:
        date_counts[row] += 1
date_counts
```

```python
sex_counts = {}
race_counts = {}

for row in data:
    if row[5] not in sex_counts:
        sex_counts[row[5]] = 1
    else:
        sex_counts[row[5]] += 1
for row in data:
    if row[7] not in race_counts:
        race_counts[row[7]] = 1
    else:
        race_counts[row[7]] += 1
```
